```python
#positive/negative simple sentiment analysis
def get_pos_neg_words():
    def get_words(url):
        import requests
        words = requests.get(url).content.decode('latin-1')
        word_list = words.split('\n')
        index = 0
        while index < len(word_list):
            word = word_list[index]
            if ';' in word or not word:
                word_list.pop(index)
            else:
                index+=1
        return word_list

    #Get lists of positive and negative words
    p_url = 'http://ptrckprry.com/course/ssd/data/positive-words.txt'
    n_url = 'http://ptrckprry.com/course/ssd/data/negative-words.txt'
    positive_words = get_words(p_url)
    negative_words = get_words(n_url)
    return positive_words,negative_words

positive_words,negative_words = get_pos_neg_words()
```

```python
from nltk import word_tokenize
```

```python
def simple_sentiment_analysis(list_name):
    ### YOUR CODE HERE ###

    positive_words, negative_words = get_pos_neg_words()
    results = list()

    for artist_name, song_name, lyrics in list_name:
        cpos = cneg = 0
        words = word_tokenize(lyrics)

        for word in words:
            if word.lower() in positive_words:
                cpos += 1
            elif word.lower() in negative_words:
                cneg += 1

        pos_ratio = round(cpos / len(words), 5)
        neg_ratio = round(cneg / len(words), 5)
        results.append((artist_name, song_name, pos_ratio, neg_ratio))

    return results
```

```python
def plot_sentiment_stacked_bar(df):
    songs = df["Song"]
    positive = df["Positive (%)"]
    negative = df["Negative (%)"]

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(songs, positive, label="Positive (%)", color="lightcoral")
    ax.bar(songs, negative, bottom=positive, label="Negative (%)", color="skyblue")

    ax.set_ylabel("Percentage (%)")
    ax.set_title("Simple Sentiment Analysis of 25 Top Rated Songs")
    ax.set_xticks(range(len(songs)))
    ax.set_xticklabels(songs, rotation=45, ha="right")
    ax.legend()

    plt.tight_layout()
    plt.show()

plot_sentiment_stacked_bar(pos_neg_table)
```

```python
def emotion_analyzer_from_lyrics(song_lyrics_list, emotion_dict):

    emotions = {x for y in emotion_dict.values() for x in y}
    results = []

    for artist_name, song_name, lyrics in song_lyrics_list:
        emotion_count = {emotion: 0 for emotion in emotions}

        total_words = len(lyrics.strip())
        for word in lyrics.split():
            if emotion_dict.get(word):
                for emotion in emotion_dict[word]:
                    emotion_count[emotion] += 1 / total_words

        # Store result
        results.append(( artist_name,song_name, emotion_count))

    return results
```

```python
# Get top 25 songs for each year (two versions had to be made due to the difference in structure)

# songs_dataset_columns = [x.columns for x in songs_datasets]
# print(songs_dataset_columns)

def get_top_songs_ver1(data, year, top_n=25):
    data = data[['track_name','artist_name','popularity']]
    top_rated = data.sort_values(by='popularity', ascending=False).head(top_n)
    top_rated = pd.DataFrame(top_rated)
    top_rated['year'] = year
    return top_rated

def get_top_songs_ver2(data, year, top_n=25):
    data = data[['Artist and Title', 'popularity']]
    data['Artist and Title'] = data['Artist and Title'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
    data['artist_name'] = data['Artist and Title'].apply(lambda x: x[0] if isinstance(x, list) else None)
    data['track_name'] = data['Artist and Title'].apply(lambda x: x[1] if isinstance(x, list) else None)
    data = data[['artist_name','track_name', 'popularity']]
    #print(data)
    top_rated = data.sort_values(by='popularity', ascending=False).head(top_n)
    top_rated = top_rated.reset_index(drop=True)

    top_rated['year'] = year
    return top_rated


# Get top songs 2018-2021
list_of_top_songs = []
for year, dataset in zip(years[:-2], songs_datasets[:-2]):
    top_songs = get_top_songs_ver1(dataset, year)
    list_of_top_songs.append(top_songs)

top_songs_2022 = get_top_songs_ver2(df_2022, 2022)
top_songs_2023 = get_top_songs_ver2(df_2023, 2023)
list_of_top_songs.append(top_songs_2022)
list_of_top_songs.append(top_songs_2023)


top_songs_df = pd.concat(list_of_top_songs, ignore_index=True)
```

```python
import lyricsgenius
import pandas as pd

def clean_lyrics(lyrics):
    """Cleans the fetched lyrics by removing unnecessary lines."""
    return "\n".join([line.strip() for line in lyrics.split("\n") if line.strip()])

def song_lyrics_all_years(songs_df, api_key='CkTeREcylNKKuBU7eLht_I7Dk16HRZzEM-3iepdtRbnDX_9waj8Rw1U2RVIPY2Ue'):
    genius = lyricsgenius.Genius(api_key)
    genius.timeout = 10  # Set timeout for requests

    songs_with_lyrics = []

    for _, row in songs_df.iterrows():
        artist_name = row['artist_name']
        song_name = row['track_name']
        year = row['year']

        try:
            song = genius.search_song(song_name, artist=artist_name)
            if song and song.lyrics:
                cleaned_lyrics = clean_lyrics(song.lyrics)
                lyrics = cleaned_lyrics
            else:
                lyrics = "Sorry, no lyrics found."
        except Exception as e:
            lyrics = f"Error: {str(e)}"

        songs_with_lyrics.append({
            "Year": year,
            "Artist": artist_name,
            "Song Title": song_name,
            "Lyrics": lyrics
        })

    return pd.DataFrame(songs_with_lyrics, columns=["Year", "Artist", "Song Title", "Lyrics"])
```
